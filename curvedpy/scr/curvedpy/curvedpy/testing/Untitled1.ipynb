```python
# a custom function that blocks for a moment
def task():
    # block for a moment
    sleep(1)
    # display a message
    print('This is from another process')
```

```python
# create a process
process = mp.Process(target=task)
```

```python
# run the process
process.start()
```

```python
def double(a):
    return a * 2

def driver_func():
    PROCESSES = 4
    with mp.Pool(PROCESSES) as pool:
        params = [(1, ), (2, ), (3, ), (4, )]
        results = [pool.apply_async(double, p) for p in params]

        for r in results:
            print('\t', r.get())
driver_func()
```
