Here we have run a 512x512 grid of geodesics for a camera "inf" far away. The coordinates of the pixels range from -50 to 50 in both directions and the inclination of the rotation axes is 89degrees (almost edge-on). The standard amount of steps of the geodesics is 250.

Now load the h5 output file:

```python
input_fname = "a0.99_th1.00_geo.h5"#a0.99_th89.00_geo.h5"#"50-50-512_a0.50_th89.00_geo.h5"
output_fname = input_fname
```

Now convert the kgeo output to a pickle file used by the Black Hole Render Engine. For this load the curvedpy package:

```python
import curvedpy as cp
cp.__version__
from curvedpy.utils.kgeo_conversions import kgeoToCurvedpy
```

```python
_ = kgeoToCurvedpy(input_fname, 512, 512, output_fname)
```
